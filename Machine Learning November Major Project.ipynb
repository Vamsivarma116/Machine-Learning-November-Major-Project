```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("adult.csv")
df1.head()
```

# 1) Renaming The Columns.

```python
"""Adding or Renaming the column Names on the top of the DataFrame"""

Attributes = ["Age","Workclass","Fnlwgt","Education","education_num","marital_status","occupation","relationship","race","sex","capital_gain","capital_loss","hours_per_week","native_country","income"]

if(Attributes!=df1.columns.values.tolist()):
    DataDict=dict(zip(Attributes,df1.columns.values.tolist()))

Adding_row = []

Adding_row.insert(0,DataDict)

df1.set_axis(["Age","Workclass","Fnlwgt","Education","education_num","marital_status","occupation","relationship","race","sex","capital_gain","capital_loss","hours_per_week","native_country","income"], axis='columns', inplace=True)

Df = pd.concat([pd.DataFrame(Adding_row), df1], ignore_index=True)

cd={"Age":int,"Fnlwgt":int,"education_num":int,"capital_gain":int,"capital_loss":int,"hours_per_week":int}

Df=Df.astype(cd)

Df
```

# 2) Removing or Handling Null Values (if any).

```python
Df.isnull().sum()
```

 """Therefore their are no null values."""

# Handling Duplicates (if any).

```python
Df.duplicated().sum()
```

"Here, We are having 24 Null values so we need to remove them."

```python
"""Handling the duplicates"""
Df.drop_duplicates(inplace=True)

print(Df.duplicated().sum()) #Checking Still we are having duplicates or not.

print(Df.shape) #Checking the shape if any changes are occured or not in the dataframe.

```

# Handling the missing values.

'''As we can see In the given adult.CSV file we are having some of the Missing values like "?" So we need to Handle them.'''

```python
Df.dtypes
```

```python
"""Checking for Missing values."""

x=[i for i in Df.columns if Df[i].dtypes == "O" ]

for feature in x:
    print(Df[feature].value_counts()) 
    print("xxxxxxxxxxxxxxxxxxxxxxxxxxx")
```

```python
"""In the above DataSet the columns having Missing Values are
    (workclass)--(occupation)--(native_country)"""


print("workclass       : ",(2093/32561)*100) #-> The percentage of Missing values in workclass.
print("occupation      : ",(1843/32561)*100) #-> The percentage of Missing values in occupation.
print("native-country  : ",(583/32561)*100)  #-> The percentage of Missing values in native-country.
```

```python
"""Here To Handle missing values we are having 3 methods 
   But, As per our DataSet the percentage of Missing values in each of the 3 columns mentioned above are very Less
   so we can replace the missing values with the value which is more frequently repeated in the Respective column."""
   #Most frequently repeated is nothing but mode of that particular column.

import warnings
warnings.filterwarnings('ignore')

Df["Workclass"] = Df['Workclass'].apply(str).str.replace('?', 'Private' )
Df['occupation'] = Df['occupation'].apply(str).str.replace('?', 'Prof-specialty' )
Df['native_country'] = Df['native_country'].apply(str).str.replace('?', 'United-States' )
```

```python
Df.head(20)
```

# Encoding The Labels (if any).

```python
"""Encoding the Label column"""

from sklearn.preprocessing import LabelEncoder

lb=LabelEncoder()

Df["income"]=lb.fit_transform(Df["income"])

Df["Workclass"]=lb.fit_transform(Df["Workclass"])

Df["Education"]=lb.fit_transform(Df["Education"])

Df["marital_status"]=lb.fit_transform(Df["marital_status"])

Df["occupation"]=lb.fit_transform(Df["occupation"])

Df["relationship"]=lb.fit_transform(Df["relationship"])

Df["race"]=lb.fit_transform(Df["race"])

Df["sex"]=lb.fit_transform(Df["sex"])

Df["native_country"]=lb.fit_transform(Df["native_country"])

Df.head(20)
```

"From the above DataFrame we can say that the required column values are Labeled Succesfully."

# 3) Spliting data into training and test data.

```python
x=Df.drop("income",axis=1) #data frame x --> independent.
y=Df["income"] #series y --> dependent.
print(type(x))
print(type(y))
print(x.shape)
print(y.shape)
```

```python
#<------spliting the data into train and test------>

from sklearn.model_selection import train_test_split

"""Spliting the Training Data and Test Data with test size of 30% from total size"""
                            
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)

x1_train,x1_test,y1_train,y1_test=train_test_split(x,y,test_size=0.80,random_state=42)

```

"""Splited the Data Succesfully."""

```python
 """Generating confusion matrix,Classification report and accuracy"""
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
def eval_model(y_test,ypred):
    print("\nConfusion Matrix : \n",confusion_matrix(y_test,ypred))
    print("\nClassification Report : \n",classification_report(y_test,ypred))
    print("\nAccuracy Score : ",accuracy_score(y_test,ypred))
def mscore(model):
    print("\nTraining Score for Training Data",model.score(x_train,y_train))
    print("\nTraining Score for Testing Data",model.score(x_test,y_test))
```

# 4,5) Applying the following models on the training dataset and generating the predicted value for the test data.


# a. Decision Tree Model ->

```python
from sklearn.tree import DecisionTreeClassifier

Dtc= DecisionTreeClassifier(criterion="gini",max_depth=40,min_samples_split=15)

Dtc.fit(x_train,y_train)

ypredDTC=Dtc.predict(x_test)

print("Predicted value for test data in Decision Tree Model Classifier is :\n",ypredDTC)

eval_model(y_test,ypredDTC)

mscore(Dtc)
```

# b. Random Forest Classifier ->

```python
from sklearn.ensemble import RandomForestClassifier

Rfc= RandomForestClassifier(n_estimators=65,criterion="gini",max_depth=28,min_samples_split=15)

Rfc.fit(x_train,y_train)

ypredRFC=Rfc.predict(x_test)

print("Predicted value for test data in Random Forest Classifier is :\n",ypredRFC)

eval_model(y_test,ypredRFC)

mscore(Rfc)
```

# c. Logistic Regression ->

```python
from sklearn.linear_model import LogisticRegression

logreg=LogisticRegression(solver='newton-cg',max_iter=1000)

logreg.fit(x_train,y_train)

ypredLR=logreg.predict(x_test)

print("Predicted value for test data in Logistic Regression is :\n",ypredLR)

eval_model(y_test,ypredLR)

mscore(logreg)
```

# d. KNN Classifier ->

```python
from sklearn.neighbors import KNeighborsClassifier

Knn=KNeighborsClassifier(n_neighbors=11)

Knn.fit(x_train,y_train)

ypredKNN=Knn.predict(x_test)

print("Predicted value for test data in KNN Classifier is :\n",ypredKNN)

eval_model(y_test,ypredKNN)

mscore(Knn)
```

# e. SVC Classifier (with linear kernel) ->

```python
from sklearn.svm import SVC

Svm=SVC(kernel="linear",C=1)

Svm.fit(x1_train,y1_train)

ypredSVM=Svm.predict(x1_test)

print("Predicted value for test data in SVC Classifier is :\n",ypredSVM)

eval_model(y1_test,ypredSVM)

mscore(Svm)
```

# 7) Validating the result for Precision, Recall, F1-score and Accuracy for each model based on values from confusion_matrix and classification_report.

# a. Decision Tree

```python
print(classification_report(y_test,ypredDTC))
x=confusion_matrix(y_test,ypredDTC)
TP=x[0][0]
TN=x[1][1]
FP=x[1][0]
FN=x[0][1]

PrecisionDTC_0 = TP/(TP+FP)

RecallDTC_0 = TP/(TP+FN)

F1ScoreDTC_0 = (2*PrecisionDTC_0*RecallDTC_0)/(PrecisionDTC_0+RecallDTC_0)

PrecisionDTC_1 = TN/(TN+FN)

RecallDTC_1 = TN/(TN+FP)

F1ScoreDTC_1 = (2*PrecisionDTC_1*RecallDTC_1)/(PrecisionDTC_1+RecallDTC_1)

AccuracyDTC = (TP+TN)/(TP+TN+FP+FN)

print("Precision '0' for Decision Tree",PrecisionDTC_0)
print("Recall    '0' for Decision Tree",RecallDTC_0)
print("F1Score   '0' for Decision Tree",F1ScoreDTC_0)
print("Precision '1' for Decision Tree",PrecisionDTC_1)
print("Recall    '1' for Decision Tree",RecallDTC_1)
print("F1Score   '1' for Decision Tree",F1ScoreDTC_1)
print("Accuracy for Decision Tree",AccuracyDTC)
```
